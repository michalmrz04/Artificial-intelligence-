Creating a rule based ai to calculate the BMI of a person

```python
def ai_response(user_input):
    # be able to read eveyrhting in lower case :p
    user_input = user_input.lower()

    # answers to user input below
    if 'hello' in user_input:
        return "Hello, how may I assist you today ?"
    elif 'hi' in user_input:
        return "Is there anything i can help you with?"
    elif 'how are you' in user_input:
        return "I am doing amazing, I was created to help you calculate your BMI. Can I help you with that right now?"
    elif 'bye' in user_input:
        return "Byeeeee! See you soon!"
    elif 'BMI' in user_input:
        return "Sure, I can help you calculate your BMI. Please provide your weight and height in meters."
    else:
        return "I'm not sure how to help you with that. Please try asking something else."
```

```python
def calc_BMI(height, weight):
    bmi = weight / (height ** 2)
    return bmi
```

```python
def run_ai():
    user_input = input(" I am here to help you calculate your BMI. To start, please type 'BMI', then provide your weight in KG and height in meters ")

    response = ai_response(user_input)
    print(response)
        
    if 'BMI' in user_input:
        weight = float(input("please enter your weight in kg"))
        height = float(input("please enter your height in meters"))
        
        bmi = calc_BMI(height, weight)
        
        print(f"Your BMI is {bmi:.2f} ")

run_ai()
```
