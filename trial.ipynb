Creating a class for my rule based ai

```python
def ai_response(user_input):
    # be able to read eveyrhting in lower case :p
    user_input = user_input.lower()

    # answers to user input below
    if 'hello' in user_input: 
        return "Hello, how may I assist you today ?"
    elif 'hi' in user_input:
        return "Is there anything i can help you with?"
    elif 'how are you' in user_input:
        return "I am doing amazing, I was created to help you calculate calories burned during your workout. Can I help you with that right now?"
    elif 'bye' in user_input:
        return "Byeeeee! See you soon!"
    elif 'calories' in user_input:
        return "To help you find the calories burned i will need to know your gender, weight in kilograms, session duration in minutes, and workout type"
    else:
        return "Sorry I am unable to answer, could you rephrase that or ask me a different question?"
```

```python

# define the function to calculate calories burned
def calc_calories(gender, weight, session_duration, workout_type):
    
    if workout_type == 'Cardio':
        total_calories = 8 * weight * session_duration * (3.5/200)
    elif workout_type == 'Strength':
        total_calories = 5 * weight * session_duration * (3.5/200)
    elif workout_type == 'Yoga':
        total_calories = 2.3 * weight * session_duration *(3.5/200) 
    else:
        total_calories = 9 * weight * session_duration * (3.5/200)
    
    # different calculation for men
    if gender == 'Male':
        total_calories = total_calories * 1.2
    else:
        total_calories = total_calories 
    return total_calories
```

```python
def use_ai():

    # ask for input 
    user_input = input("I am an AI designed to help you find the amount of calories you burned. Type 'calories' to get started! ")

    # checks for user input of 'calories'
    if 'calories' in user_input.lower():
        # ask for the user details to complete the prediction/ calculation
        gender = input("Are you a Male or Female?")
        weight = float(input("What is your weight in kg?"))
        session_duration = float(input("How long was your workout session in minutes?"))
        workout_type = input("What did you do? Did you do Cardio, HIIT, Yoga, or Strength?")

        # calculate calories burned
        calories = calc_calories(gender, weight, session_duration, workout_type)
        # print the calories burned to two decimal places
        print(f"Calories burned: {calories:.2f}")
    else:
         # Respond to user input 
        response = ai_response(user_input)
        print(response)

# run the ai 
use_ai()
```
